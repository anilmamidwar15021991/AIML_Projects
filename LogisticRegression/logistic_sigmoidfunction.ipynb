```python
import pandas as pd
import numpy as np
from numpy import random
```

```python
random1=[]
random2=[]
random3=[]
random4=[]
xlist=[]
   
theta=[10,1,0.1,-1]
```

```python
#for same value of X with different theta value will show different s curves
import math
for i in range(100):
    x=np.random.uniform(-5,5)
    xlist.append(x)
    logreg1=1/(1+ math.exp(-(theta[0]*x)))
    logreg2=1/(1+ math.exp(-(theta[1]*x)))
    logreg3=1/(1+ math.exp(-(theta[2]*x)))
    logreg4=1/(1+ math.exp(-(theta[3]*x)))
    random1.append(logreg1)
    random2.append(logreg2)
    random3.append(logreg3)
    random4.append(logreg4)

    
```

```python
import matplotlib.pyplot as plt

plt.scatter(xlist,random1,marker='*',s=40,c='orange',alpha=0.5,label=r'$\theta = %3.1f$'%(theta[0]))
plt.scatter(xlist,random2,marker='*',s=40,c='magenta',alpha=0.5,label=r'$\theta = %3.1f$'%(theta[1]))
plt.scatter(xlist,random3,marker='*',s=40,c='green',alpha=0.5,label=r'$\theta = %3.1f$'%(theta[2]))
plt.scatter(xlist,random4,marker='*',s=40,c='red',alpha=0.5,label=r'$\theta = %3.1f$'%(theta[3]))
plt.axhline(y=0.5,label='P=0.5')
plt.legend(fontsize=16)
plt.show()

#higher the theta value more s chaped curve will be present, with a=0 
```

```python
random11=[]
random12=[]
random13=[]
random14=[]
xlist1=[]
a=[-10,1,10]

for i in range(100):
    y= np.random.uniform(-5,5)
    xlist1.append(y)
    logreg11=1/(1+ math.exp(-(a[0]+(10*y))))
    logreg12=1/(1+ math.exp(-(a[1]+(10*y))))
    logreg13=1/(1+ math.exp(-(a[2]+(10*y))))
    random11.append(logreg11)
    random12.append(logreg12)
    random13.append(logreg13)
    
    
```

```python
plt.scatter(xlist1,random11,marker='*',s=40,c='orange',alpha=0.5)
plt.scatter(xlist1,random12,marker='*',s=40,c='red',alpha=0.5)
plt.scatter(xlist1,random13,marker='*',s=40,c='blue',alpha=0.5)
plt.axhline(y=0.5,label='P=0.5')
plt.legend(fontsize=16)
plt.show()

#this show if we increase value of a . it will shift the s curve along the x-axis*
```
