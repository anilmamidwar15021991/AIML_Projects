```python
df_afterdummies_limitedfeatures = df_afterdummies[['satisfaction_level','average_montly_hours','promotion_last_5years','Sal_high','Sal_low','Sal_medium','left']]

df_afterdummies_limitedfeatures
```

```python
from imblearn.over_sampling import RandomOverSampler
X = df_afterdummies_limitedfeatures.drop(['left'],axis=1)
y= df_afterdummies_limitedfeatures['left']
Sampler = RandomOverSampler()
X_sampled,y_sampled = Sampler.fit_resample(X,y)
```

```python
#EDA completed above now will do model building scaling all the features to one level between 0 & 1
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
sc = StandardScaler()
X_train,X_test,y_train,y_test = train_test_split(X_sampled,y_sampled,train_size=0.3)
X_train=sc.fit_transform( X_train )
X_test=sc.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X_train,y_train)
```

```python
y_predicted = LR.predict(X_test)
y_predictedProba = LR.predict_proba(X_test)
```
